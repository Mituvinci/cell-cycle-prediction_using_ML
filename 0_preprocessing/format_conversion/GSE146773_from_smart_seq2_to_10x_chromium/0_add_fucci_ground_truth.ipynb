```python
import os
import pandas as pd
import gzip

# Set the directory where the .txt.gz files are located
directory_path = "D:/Halima's Data/Thesis_2/RCode/GSE146773_RAW/" # Replace with your actual directory path


# Initialize an empty DataFrame to store the combined data
combined_df = None

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.txt.gz'):
        # Construct the full file path
        file_path = os.path.join(directory_path, filename)
        # Read the contents of the .txt.gz file into a DataFrame
        with gzip.open(file_path, 'rt') as file:
            # Assuming the file has a header and the gene_id is in the first column
            df = pd.read_csv(file, sep='\t', usecols=['gene_id', 'expected_count'])
            # Extract the gene ID after 'gene:'
            df['gene_id'] = df['gene_id'].str.split(':').str.get(1)

            
            # Rename 'expected_count' with the cell name extracted from the filename
            cell_name = filename.split('_')[2]+"_"+filename.split('_')[1]  # Adjust index as needed based on filename format
            #print(cell_name)
            
            df.rename(columns={'expected_count': cell_name}, inplace=True)
            #print("second")
            #print(df.head())
            # Set the gene_id as the index
            df.set_index('gene_id', inplace=True)

            #print("third")
            #print(df.head())

            
            # Transpose the DataFrame to have gene_ids as columns and the cell name as the row
            df = df.T

            #print(df.head())

            
            # If combined_df is None, initialize it with the columns of the first df
            if combined_df is None:
                combined_df = pd.DataFrame(columns=df.columns)
            # Append to the combined DataFrame
            combined_df = pd.concat([combined_df, df])

            #print(combined_df.head())
            

# Reset index to make the cell names a column again
combined_df.reset_index(inplace=True)
# Rename the column to 'CellID'
combined_df.rename(columns={'index': 'CellID'}, inplace=True)


# Save the combined DataFrame to a CSV file
output_file_path = "D:/Halima's Data/Thesis_2/RCode/GSE146773_RAW_final_output.csv"  # Replace with your desired output path
combined_df.to_csv(output_file_path, index=False)

# Print the path to the saved CSV file
print(f"Combined data saved to {output_file_path}")
```
