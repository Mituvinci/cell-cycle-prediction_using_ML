```python

# Load the data from your CSV file
import scanpy as sc
import pandas as pd
import anndata

# Load the TPM file
tpm_file = "D:/Halima's Data/Thesis_2/RCode/Cell_Cycle_prediction_with_scATAC_Seq/paper1/GSE64016_H1andFUCCI/GSE64016_H1andFUCCI_normalized_EC.csv"
tpm_data = pd.read_csv(tpm_file, index_col=0)  # Gene names as row indices

# Check the shape of tpm_data
print(f"tpm_data shape: {tpm_data.shape}")  # Should be (genes, cells)

# Verify that genes are rows and cells are columns
if tpm_data.shape[0] < tpm_data.shape[1]:
    raise ValueError("The data matrix appears transposed. Ensure genes are rows and cells are columns.")

# Create AnnData object
adata = anndata.AnnData(X=tpm_data.T)  # Transpose for AnnData if needed
adata.var_names = tpm_data.index       # Set gene names as variable names
adata.obs_names = tpm_data.columns     # Set cell names as observation names

# Confirm AnnData structure
print(adata)


```

```python
adata
```

```python
# Predict cell cycle phases
predicted_labels, probabilities = ccAFv2.predict_labels(adata, species='human', gene_id='symbol')

# Add predictions to the AnnData object
adata.obs['ccAFv2'] = predicted_labels
```
